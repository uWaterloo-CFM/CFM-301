```python
import pandas as pd
from pandasql import sqldf
```

```python
# Goal: Append quarterly financials (compustat) to monthly data (crsp) 
crsp = pd.read_sas("crsp_nasdaq100.sas7bdat", encoding= "ISO-8859-1 ") # monthly, starts 2020
compustat = pd.read_sas("compustat_nasdaq100.sas7bdat", encoding= "ISO-8859-1") #quarterly, starts 2019
```

```python
sqldf("SELECT * FROM crsp LIMIT 5")
```

```python
sqldf("SELECT * FROM compustat LIMIT 5")
```

```python
crsp['YEAR'] = pd.DatetimeIndex(crsp['DATE']).year
crsp['MONTH'] = pd.DatetimeIndex(crsp['DATE']).month
crsp['monthid'] = ((crsp['YEAR']) - 2000) * 12 + crsp['MONTH']
sqldf("SELECT * FROM crsp LIMIT 5")
```

```python
compustat['YEAR'] = pd.DatetimeIndex(compustat['DATADATE']).year
compustat['MONTH'] = pd.DatetimeIndex(compustat['DATADATE']).month
compustat['monthid'] = ((compustat['YEAR']) - 2000) * 12 + compustat['MONTH']
sqldf("SELECT * FROM compustat LIMIT 5")
```

```python
df_compustat_crsp = sqldf("SELECT a.*, b.* \
                           FROM crsp as a \
                           LEFT JOIN compustat as b \
                           ON a.TICKER = b.TIC and a.monthid >= b.monthid + 4 and a.monthid <= b.monthid + 6")
```
