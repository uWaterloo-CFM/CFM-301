```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load crsp data
ff_factors = pd.read_sas("farma-french.sas7bdat", encoding = 'latin-1')
```

```python
ff_factors[:250]
```

```python
plt.hist(ff_factors["SMB"], bins=50)
plt.show()
```

```python
plt.hist(ff_factors["HML"], bins=50)
plt.show()
```

```python
plt.hist(ff_factors["UMD"], bins=50) # MOM
plt.show()
```

```python
from scipy import stats
import numpy as np
print("SMB: ", stats.ttest_ind(ff_factors["SMB"], np.zeros(len(ff_factors["UMD"]))))
print("HML: ", stats.ttest_ind(ff_factors["HML"], np.zeros(len(ff_factors["UMD"]))))
print("UMD: ", stats.ttest_ind(ff_factors["UMD"], np.zeros(len(ff_factors["UMD"]))))
stats.ttest_ind(ff_factors["UMD"], np.zeros(len(ff_factors["UMD"])))[0]
```

## Conclusions
So SMB is not significantly different than zero

HML is significantly different than zero

UMD is significantly different than zero at a very strong level of confidence
